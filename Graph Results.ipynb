Upload excel file - these values were taken from the results obtained from the actual analysis. The data file contains the chemical name, current UK spending in pounds, and projected UK spending in pounds.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

data = pd.read_excel("values.xlsx")
```

We sort the data here by total spending for the purposes of making a clean graph

```python
data = data.sort_values(by='Projected UK Spending in Pounds', ascending=False)
```

```python
data.head(5)
```

We graph data for actual and projected UK spending in USD for each chemical.

```python
plt.rc('font', size=10)         
plt.rc('axes', titlesize=10)   
plt.rc('axes', labelsize=15)   
plt.rc('xtick', labelsize=15)  
plt.rc('ytick', labelsize=12)    
plt.rc('legend', fontsize=15)   
plt.rcParams["font.family"] = "sans-serif"


ind = np.arange(len(data))
width = 0.4

fig, ax = plt.subplots(figsize = (18, 18))
ax.barh(ind, data['Projected UK Spending in Pounds'], width, color=(0.2, 0.4, 0.6, 0.6), label='2018 Projected English Spending')
ax.barh(ind + width, data['sum_cost_UK_pounds'], width, color=(0, 0, 0.8, 1.0), label='2018 English Spending')
ax.xaxis.grid(True, linestyle='--', which='major',
                   color='grey', alpha=.25)
opacity = 0.8

ax.set(yticks= ind + width, yticklabels=data['chemical'], ylim=[2*width - 1, len(data)])
ax.legend()
ax.set_xlabel('Spending,  (100 million)')

plt.savefig('UK_US_price.jpeg',dpi=600, bbox_inches = "tight")
```
